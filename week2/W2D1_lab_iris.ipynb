### الخطوة 1 — حمّل البيانات وانظر إليها
كم عيّنة عندنا؟ وما القياسات المتاحة؟

</div>

```python
import pandas as pd

# البيانات تُحمَّل من الإنترنت مباشرة، لا حاجة لرفع أي ملف
# هذه الأسطر تجعل التحميل يعمل على أي منصة، شغّلها ولا تقلق من تفاصيلها
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
try:
    from pyodide.http import open_url
    df = pd.read_csv(open_url(url))
except ImportError:
    df = pd.read_csv(url)

print(df.head())
print("Number of flowers:", len(df))
print("Columns:", list(df.columns))
```

<div dir="rtl" align="right">

### الخطوة 2 — خذ عمودين كقائمتين
الآن نعمل عليهما بمهارات الأسبوع الأول: القوائم والحلقات والشروط.

</div>

```python
petal_length = df["petal_length"].tolist()
species = df["species"].tolist()

print(petal_length[0], species[0])
```

<div dir="rtl" align="right">

### الخطوة 3 — عُدّ كل نوع
هل المجموعة متوازنة؟

</div>

```python
setosa = 0
versicolor = 0
virginica = 0

for s in species:
    if s == "setosa":
        setosa = setosa + 1
    elif s == "versicolor":
        versicolor = versicolor + 1
    else:
        virginica = virginica + 1

print("setosa:", setosa)
print("versicolor:", versicolor)
print("virginica:", virginica)
```

<div dir="rtl" align="right">

### الخطوة 4 — إحصاءاتٌ بسيطة
نلخّص العمود كلّه في ثلاثة أرقام: المتوسّط، وأقصر بتلة، وأطول بتلة.

› المدى واسع. لماذا في رأيك؟

</div>

```python
print("Average:", sum(petal_length) / len(petal_length))
print("Shortest:", min(petal_length))
print("Longest:", max(petal_length))
```

<div dir="rtl" align="right">

### الخطوة 5 — متوسّطٌ لكل نوع
بدل متوسّطٍ واحد لكل الأزهار، نحسب متوسّطًا لكل نوعٍ على حِدة.

› **هنا يظهر الاكتشاف.** ما الذي تلاحظه؟

</div>

```python
for name in ["setosa", "versicolor", "virginica"]:
    total = 0
    count = 0
    for i in range(len(species)):
        if species[i] == name:
            total = total + petal_length[i]
            count = count + 1
    print(name, "average petal:", total / count)
```

<div dir="rtl" align="right">

### الخطوة 6 — المدى لكل نوع
أصغر وأكبر طول بتلةٍ في كل نوع.

› أيّ نوعين يتداخلان؟

</div>

```python
for name in ["setosa", "versicolor", "virginica"]:
    picked = []
    for i in range(len(species)):
        if species[i] == name:
            picked.append(petal_length[i])
    print(name, "from", min(picked), "to", max(picked))
```
